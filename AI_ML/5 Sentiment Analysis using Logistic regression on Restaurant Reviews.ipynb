```python
import re
import nltk
# nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer

corpus = []
for i in range(0,1000):
    review = re.sub('[^a-zA-Z]',' ',dataset['Review'][i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ ps.stem(word) for word in review if not word in set(stopwords.words('english')) ]
    review = ' '.join(review)
    print(review)
    corpus.append(review)
```

```python
type(review)
```

```python
corpus
```

```python
corpus_dataset = pd.DataFrame(corpus)
corpus_dataset.head()
```

```python
corpus_dataset['corpus'] = corpus_dataset
corpus_dataset.head()
```

```python
corpus_dataset = corpus_dataset.drop([0], axis=1)
corpus_dataset.head()
```

```python
corpus_dataset.to_csv("corpus_dataset.csv")
```

```python
# Create a Bag of Words Model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1500)
```

```python
X = cv.fit_transform(corpus).toarray()
X[0]
```

```python
len(X[0])
```

```python
cv.get_feature_names_out()
```

```python
len(cv.get_feature_names_out())
```

```python
y = dataset.iloc[:,1].values
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
Review = "nice service"
input1 = [Review]

input_data = cv.transform(input1).toarray()

input_pred = classifier.predict(input_data)

if input_pred[0]==1:
    print("Review is Positive")
else:
    print("Review is Negative")
```

```python
Review = "long waiting time"
input1 = [Review]

input_data = cv.transform(input1).toarray()

input_pred = classifier.predict(input_data)

if input_pred[0]==1:
    print("Review is Positive")
else:
    print("Review is Negative")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

for c in [0.01, 0.05, 0.25, 0.5, 1]:

    lr = LogisticRegression(penalty='l2', C=c, max_iter=7600, random_state=42)
    lr.fit(X_train, y_train)
    print ("Accuracy for C=%s: %s"
           % (c, accuracy_score(y_test, lr.predict(X_test))))
```

```python
final_model = LogisticRegression(penalty='l2', C=0.25, max_iter=7600, random_state=42)
final_model.fit(X_train, y_train)
print ("Final Accuracy: %s"
       % accuracy_score(y_test, final_model.predict(X_test)))
```

```python
feature_to_coef = {
    word: coef for word, coef in zip(
        cv.get_feature_names_out(), final_model.coef_[0]
    )
}
print("POSITIVE")
for best_positive in sorted(
    feature_to_coef.items(),
    key=lambda x: x[1],
    reverse=True)[:5]:
    print (best_positive)
print("NEGATIVE")
for best_negative in sorted(
    feature_to_coef.items(),
    key=lambda x: x[1])[:5]:
    print (best_negative)
```

```python
y_pred_lr = final_model.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred_lr)
```

```python
accuracy_score(y_test,y_pred_lr)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```
