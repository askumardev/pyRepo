```python
# Divide the dataset into features and response
X = df.drop(['sales'], axis=1) # Features
y = df['sales'] # Response
```

```python
X.head()
```

```python
y.head()
```

```python
# https://scikit-learn.org/1.5/modules/generated/sklearn.model_selection.train_test_split.html
from sklearn.model_selection import train_test_split # Import the train_test_split function
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# random_state=0 specifies that the random number generator is the RandomState instance
# test_size=0.2 specifies that 20% of the data is to be used for testing
# random_state can take any integer value. It is used to reproduce the same results across different runs
# The value that random_state takes ranges from 0 to 4294967296 i.e. 0 to 2^32-1
# Some popular values are 0, 1, 42, 100, 123. These values are used to reproduce the same results across different runs

X_train.head()
```

```python
X_train.describe()
```

```python
features = ['TV', 'radio', 'newspaper'] # List of features
features
```

```python
# MinMax Scaling
# https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html
# MinMaxScaler scales the data to a fixed range [0, 1]
from sklearn.preprocessing import MinMaxScaler # Import the MinMaxScaler class
mms = MinMaxScaler() # Create an instance of the MinMaxScaler class
X_train[features] = mms.fit_transform(X_train[features]) # fit means studying the data, transform means change the data in the range [0, 1]
X_test[features] = mms.transform(X_test[features]) # transform means change the data in the range [0, 1]
```

```python
X_train.describe()
```

```python
# Lets create a Linear Regression model
# https://scikit-learn.org/1.5/modules/generated/sklearn.linear_model.LinearRegression.html
from sklearn.linear_model import LinearRegression # Import the LinearRegression class
lr = LinearRegression() # Create an instance of the LinearRegression class
lr.fit(X_train, y_train) # # fit means studying the data
y_pred_train = lr.predict(X_train) # predict the response on the training data
y_pred_test = lr.predict(X_test) # predict the response on the testing data
from sklearn.metrics import r2_score # Import the r2_score function. This will give value beween [0, 1], higher the better. 0.8 is good
train_accuracy = r2_score(y_train, y_pred_train) # Calculate the R2 score on the training data
test_accuracy = r2_score(y_test, y_pred_test) # Calculate the R2 score on the testing data
print("Training Accuracy: ", train_accuracy) # Print the training accuracy
print("Testing Accuracy: ", test_accuracy) # Print the testing accuracy
```

```python
# Lets create a Support Vector Regression model
# https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVR.html
from sklearn.svm import SVR # Import the SVR class
svr = SVR() # Create an instance of the SVR class
svr.fit(X_train, y_train) # fit means studying the data
y_pred_train = svr.predict(X_train) # predict the response on the training data
y_pred_test = svr.predict(X_test) # predict the response on the testing data
train_accuracy = r2_score(y_train, y_pred_train) # Calculate the R2 score on the training data
test_accuracy = r2_score(y_test, y_pred_test) # Calculate the R2 score on the testing data
print("Training Accuracy: ", train_accuracy) # Print the training accuracy
print("Testing Accuracy: ", test_accuracy) # Print the testing accuracy
```

```python
# Qunatify what is a good model:
# (Train Accuracy - Test Accuracy) <= 0.05
# Train Accuracy >= 0.8
# If both the above conditions are satisfied, then the model is a good model
```

```python
# Lets create a AdaBoost Regression model
# https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostRegressor.html
from sklearn.ensemble import AdaBoostRegressor # Import the AdaBoostRegressor class
abr = AdaBoostRegressor() # Create an instance of the AdaBoostRegressor class
abr.fit(X_train, y_train) # fit means studying the data
y_pred_train = abr.predict(X_train) # predict the response on the training data
y_pred_test = abr.predict(X_test) # predict the response on the testing data
train_accuracy = r2_score(y_train, y_pred_train) # Calculate the R2 score on the training data
test_accuracy = r2_score(y_test, y_pred_test) # Calculate the R2 score on the testing data
print("Training Accuracy: ", train_accuracy) # Print the training accuracy
print("Testing Accuracy: ", test_accuracy) # Print the testing accuracy
```

```python
# Lets implement the CatBoost Regressor
# Install the catboost library. -q means quiet mode
!pip install catboost -q
from catboost import CatBoostRegressor # Import the CatBoostRegressor class
cbr = CatBoostRegressor(depth=2, iterations=500, learning_rate=0.1) # Create an instance of the CatBoostRegressor class
cbr.fit(X_train, y_train) # fit means studying the data
y_pred_train = cbr.predict(X_train) # predict the response on the training data
y_pred_test = cbr.predict(X_test) # predict the response on the testing data
train_accuracy = r2_score(y_train, y_pred_train) # Calculate the R2 score on the training data
test_accuracy = r2_score(y_test, y_pred_test) # Calculate the R2 score on the testing data
print("Training Accuracy: ", train_accuracy) # Print the training accuracy
print("Testing Accuracy: ", test_accuracy) # Print the testing accuracy
```
